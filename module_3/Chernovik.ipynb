```python
def splitting_row(str_val):
    if pd.isna(str_val):
        return ["Unknown"]
    str_val = str_val.strip('[]')
    str_val = str_val.replace("\'",'')
    str_val = str_val.split(",")
    return str_val
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('main_task.xls')
```

```python
def splitting_row(str_val):
    if pd.isna(str_val):
        return ["Unknown"]
    str_val = str_val.strip('[]')
    str_val = str_val.replace("\'",'')
    str_val = str_val.split(",")
    return str_val
```

```python
df['Cuisine Style'] = df['Cuisine Style'].apply(splitting_row)
```

```python
df['Cuisine Style'][0]
```

```python
def counter(row):
    for i in row:
        if i=='Unknown':
            return 1
        else:
            return len(i)
```

```python
df['Cuisine Style']= df['Cuisine Style'].apply(counter)
```
